```python
import os
import json
import pandas as pd
```

```python
directory = '../src/experiments/results'
all_results = []
for file in os.listdir(directory):
    if file.endswith('pkl'):
        continue
    
    data_type = 'balanced' if file.startswith('balanced') else 'imbalanced'
    
    metrics=json.load(open(os.path.join(directory, file), 'r'))
    
    train_metrics = metrics['metrics']['train']
    train_metrics = {'train_{}'.format(key): train_metrics[key] for key in train_metrics}
    
    test_metrics = metrics['metrics']['test']
    test_metrics = {'test_{}'.format(key): test_metrics[key] for key in test_metrics}
    
    metrics.update(train_metrics)
    metrics.update(test_metrics)
    
    metrics['metrics'] = None
    metrics['data_type'] = data_type
    
    metrics['file'] = file
    
    all_results.append(metrics)
```

```python
df = pd.DataFrame(all_results)
```

```python
df.loc[df.data_type == 'imbalanced'].sort_values(['test_f1_macro', 'test_accuracy'], ascending=False)
```
